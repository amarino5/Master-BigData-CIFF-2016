Funcion lambda para la trnasformacion de los timestamps en fechas

```python
changeToDate = lambda x: datetime.datetime.utcfromtimestamp(x).strftime('%d/%m/%Y')
```

Lectura de cada fichero con los datos de la fluctuacion del bitcoin por moneda  

```python

dfCAD = pd.read_csv("data/coinbaseCAD.csv") 
dfCAD.columns=['date','price','amount'] 
dfCAD["exchange"] = "CAD" 
dfCAD['date']=dfCAD['date'].apply(changeToDate)
dfCAD.head(5)
```

Cargamos el resto de ficheros y los añadimos al dataset haciendo un merge cou outer para que meta todas las 
variaciones de todas las fechas 

```python
dfEUR = pd.read_csv("data/coinbaseEUR.csv")
dfEUR.columns=['date','price','amount'] 
dfEUR["exchange"] = "EUR" 
dfEUR['date']=dfEUR['date'].apply(changeToDate)

dfGBP = pd.read_csv("data/coinbaseGBP.csv") 
dfGBP.columns=['date','price','amount']
dfGBP["exchange"] = "GBP" 
dfGBP['date']=dfGBP["date"].apply(changeToDate)

dfTotal = dfCAD.merge(dfEUR.merge(dfGBP,how='outer'),how='outer')
#dfTotal = dfB7.merge(dfBitalo,how='outer')
dfTotal.head(5)
```

establecer como indice la fecha de la medicion

```python
#dfTotal.reset_index()
dfTotal= dfTotal.set_index(['date'])
dfTotal= dfTotal.dropna()
dfTotal.head(5)
```

```python
#guardar a excel por si fuese necesario comprobar resultados
dfTotal.to_csv('dfTotal.csv')
```

Calculamos los Spreads para cada exchange para cada dia, de cara a obtener el mayor

```python
dfTotal_noidx= dfTotal.reset_index()
```

```python
SspreadsCAD = dfTotal_noidx[dfTotal_noidx['exchange']=='CAD'].groupby(['date'])['price'].max() - dfTotal_noidx[dfTotal_noidx['exchange']=='CAD'].groupby(['date'])['price'].min()
SspreadsEUR = dfTotal_noidx[dfTotal_noidx['exchange']=='EUR'].groupby(['date'])['price'].max() - dfTotal_noidx[dfTotal_noidx['exchange']=='EUR'].groupby(['date'])['price'].min()
SspreadsGBP = dfTotal_noidx[dfTotal_noidx['exchange']=='GBP'].groupby(['date'])['price'].max() - dfTotal_noidx[dfTotal_noidx['exchange']=='GBP'].groupby(['date'])['price'].min()
```

```python
dfSpreadCAD = pd.DataFrame({'spreadCAD':SspreadsCAD})
dfSpreadCAD_noindex = dfSpreadCAD.reset_index()
dfSpreadEUR = pd.DataFrame({'spreadEUR':SspreadsEUR})
dfSpreadEUR_noindex= dfSpreadEUR.reset_index()
dfSpreadGBP = pd.DataFrame({'spreadGBP':SspreadsGBP})
dfSpreadGBP_noindex= dfSpreadGBP.reset_index()
dfMerged = dfSpreadCAD_noindex.merge(dfSpreadEUR_noindex, how='inner')
dfMerged = dfMerged.merge(dfSpreadGBP_noindex, how='inner')
dfMerged = dfMerged.set_index(['date'])
dfMerged.head(5)
```
